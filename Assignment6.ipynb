```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Iris.csv")
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Species"].value_counts()
```

```python
df = df.drop('Id', axis=1)
df.head()
```

```python
corr = df.corr()
sns.heatmap(corr, annot= True, cmap='coolwarm')
```

```python
X=df.drop('Species',axis=1)
y=df['Species']
X
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state=430)
```

```python
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X,y)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
predicted_dataframe = pd.DataFrame({"Predicted: ":y_pred, "Actual: ": y_test})
print("Predicted and Actual  Data frame")
print(predicted_dataframe)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(conf_matrix)

TP = conf_matrix[1][1]
FP = conf_matrix[0][1]
TN = conf_matrix[0][0]
FN = conf_matrix[1][0]

print("TP: ", TP)
print("FP: ", FP)
print("FN: ", FN)
print("TN: ", TN)

# accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

#  error rate
error_rate = (FP + FN) / (TP + TN + FP + FN)

#  precision
precision = TP / (TP + FP)

#  recall
recall = TP / (TP + FN)

print("Accuracy:", accuracy)
print("Error Rate:", error_rate)
print("Precision:", precision)
print("Recall:", recall)
```
