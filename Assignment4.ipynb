```python
X = df.drop(columns=["medv","rad"], axis=1)
y = df["medv"]
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
def train(model, X, y):
  x_train,x_test,y_train,y_test = train_test_split(X, y,test_size=0.33, random_state = 42)
  model.fit(x_train, y_train)


  pred = model.predict(x_test)

  cv_score = cross_val_score(model, X,y, scoring="neg_mean_squared_error", cv=5)
  cv_score = np.abs(cv_score)

  print("Model Report")
  print("MSE: ", mean_squared_error(y_test, pred))
  print("CV Score: ", cv_score) 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
train(model, X, y)
```

```python
predicted_price_y_test = model.predict(X_test)
predicted_price_y_train = model.predict(X_train)
```

```python
model.score(X_test,y_test)
```

```python

predicted_dataframe = pd.DataFrame({"Predicted_Prices": predicted_price_y_test, "Actual_Prices": y_test})
print("Predicted and Actual Price Data frame")
print(predicted_dataframe)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("Testing: ")
rmse = mean_squared_error(y_test, predicted_price_y_test, squared=False)
print("RMSE:", rmse)

# Calculating R-squared score
r2 = r2_score(y_test, predicted_price_y_test)
print("R-squared:", r2)
print()

print("Training: ")
rmse = mean_squared_error(y_train, predicted_price_y_train, squared=False)
print("RMSE:", rmse)

# Calculating R-squared score
r2 = r2_score(y_train, predicted_price_y_train)
print("R-squared:", r2)
```



> If there isn't much difference between the R2 score of the training and test sets, and the R2 score is close to 1, it suggests that the model may not be overfitted. However, it does not provide conclusive evidence on its own. Additional analysis and evaluation methods are required to determine if the model is overfitted.
