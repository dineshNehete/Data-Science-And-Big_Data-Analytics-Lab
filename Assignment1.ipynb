```python
df['Benefits'] = df['Benefits'].replace('Not Provided', np.nan)
df['Benefits'] = df['Benefits'].astype(float)
print(df['Benefits'].unique())
```

```python
df["BasePay"].unique()
```

```python
#Cleaning base pay -> converting not provided into missing values and rest of string values into float values
df["BasePay"].replace('Not Provided', np.nan, inplace =True)
df["BasePay"] = df["BasePay"].astype(float)
print(df["BasePay"].unique())
```

```python
df["BasePay"].unique()
```

```python
#Cleaning overitme pay -> converting not provided into missing values and rest of string values into float values
df["OvertimePay"].unique()
df["OvertimePay"].replace('Not Provided', np.nan, inplace= True)
df["OvertimePay"] = df["OvertimePay"].astype(float)
df["OvertimePay"].unique()
```

```python
df["OtherPay"].unique()
```

```python
#Cleaning other pay -> converting not provided into missing values and rest of string values into float values
df["OtherPay"].replace('Not Provided', np.nan, inplace=True)
df["OtherPay"] = df["OtherPay"].astype(float)
df["OtherPay"].unique()
```

```python
df["Agency"].unique()
```

```python
df["Agency"] = df["Agency"].astype('|S')
df["Agency"]
```

```python
df["BasePay"].unique()
```

```python
#dropping all the records with missing values and replacing the 
df.dropna(subset=['BasePay'], inplace=True)
df.isnull().sum()
```

```python
df.dropna(subset=['Benefits'], inplace=True)
# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical variable
df['Status'] = label_encoder.fit_transform(df['Status'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
column_values = df['BasePay'].values.reshape(-1, 1)
normalized_values = scaler.fit_transform(column_values)
df['NormalizedBasePay'] = normalized_values
```

```python
df
```

```python
overtime_pay_min = df["OvertimePay"].min()
overtime_pay_max = df["OvertimePay"].max()

df["NormalizedOvertimePay"] = (df["OvertimePay"] - overtime_pay_min) / (overtime_pay_max - overtime_pay_min)
df
```

```python
# Z-score normalization technique->(x - mean) / std

benefits_mean = df["Benefits"].mean()
benefits_std = df["Benefits"].std()

df["Normalized_Benefits"] = (df["Benefits"] - benefits_mean) / benefits_std
df
```
